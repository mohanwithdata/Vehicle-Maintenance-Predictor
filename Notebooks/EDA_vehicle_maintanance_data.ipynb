```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("vehicle_maintenance_data.csv")
data.head(3)
```

```python
vehicle_count = pd.pivot_table(data, index='Vehicle_Model', aggfunc='size').reset_index(name='count')
```

```python
vehicle_count
```

```python
vehicle_fueltype_count = pd.pivot_table(data, index='Vehicle_Model', columns='Fuel_Type', aggfunc='size')
```

```python
vehicle_fueltype_count
```

```python
vehicle_accident_count = pd.pivot_table(data, index='Vehicle_Model', values='Accident_History', aggfunc='sum')
```

```python
vehicle_accident_count
```

```python
vehicle_transmission_count = pd.pivot_table(data, index='Vehicle_Model', columns='Transmission_Type', aggfunc='size')
```
